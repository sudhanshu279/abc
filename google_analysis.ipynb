```python
import pandas as pd
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

# PyTrends Setup aur Keyword Define Karna:

```python
pytrends = TrendReq(hl='en-US', tz=360)
keyword = "cloud computing"
```

# Data Request Karna

```python
pytrends.build_payload([keyword], cat=0, timeframe='today 12-m', geo='' , gprop='')
```

# Country-wise Interest Nikalna

```python
region_data = pytrends.interest_by_region()
region_data = region_data.sort_values(by=keyword, ascending=False).head(15)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=region_data[keyword], y=region_data.index, palette='Blues_d')
plt.title(f"Top Countries Searching '{keyword}'")
plt.xlabel("Interest")
plt.ylabel("Country")
plt.show()
```

# Choropleth (World Map Plot):

```python
region_data = region_data.reset_index()
fig = px.choropleth(region_data,
locations='geoName',
locationmode='country names',
color=keyword,
title=f"Search Interest for '{keyword}' by Country",
color_continuous_scale='Blues')
fig.show()
```

# Time-wise Interest Nikalna:

```python
time_df = pytrends.interest_over_time()
```

# Line Plot (Search Trend Over Time)

```python
plt.figure(figsize=(12,6))
plt.plot(time_df.index, time_df[keyword], marker='o', color='purple')
plt.title(f"Search Interest Over Time for '{keyword}'")
plt.xlabel("Date")
plt.ylabel("Interest")
plt.grid(True)
plt.xticks(rotation=45)
```

# Multiple Keywords Compare Karna

```python
kw_list = ["cloud computing", "data science", "machine learning"]
pytrends.build_payload(kw_list, cat=0, timeframe='today 12-m', geo=
''
, gprop=
'')
```

# Comparison Plot (Line Plot of All 3 Keywords)

```python
compare_df = pytrends.interest_over_time()

plt.figure(figsize=(12, 6))

for kw in kw_list:
    plt.plot(compare_df.index, compare_df[kw], label=kw)

plt.title("Keyword Comparison Over Time")
plt.xlabel("Date")
plt.ylabel("Search Interest")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
